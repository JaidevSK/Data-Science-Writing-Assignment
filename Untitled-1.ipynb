```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the datasets
crop_yield_df = pd.read_csv('crop_yield.csv')
rain_df = pd.read_csv('Sub_Division_IMD_2017.csv')
temp_df = pd.read_csv('TEMP_ANNUAL_SEASONAL_MEAN.csv')
```

```python

# Initial exploration of crop_yield
print("Crop Yield Data:")
print(crop_yield_df.head())
print(crop_yield_df.info())
print(crop_yield_df.describe())
print(crop_yield_df['State'].unique())
print(crop_yield_df['Crop'].unique())

# Plotting the initial exploration
plt.figure()
crop_yield_df['Yield'].hist()
plt.title('Yield Histogram')
plt.xlabel('Yield')
plt.ylabel('Frequency')
plt.show()

# Types of crops
crop_yield_df['Crop'].value_counts().plot(kind='bar')
plt.title('Crops')
plt.xlabel('Crop')
plt.ylabel('Frequency')
plt.show()

# Types of states
crop_yield_df['State'].value_counts().plot(kind='bar')
plt.title('States')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()


```

```python
# Initial exploration of rain
print("Rain Data:")
print(rain_df.head())
print(rain_df.info())
print(rain_df.describe())
print(rain_df['SUBDIVISION'].unique())

# Plotting the initial exploration
plt.figure()
rain_df['ANNUAL'].hist()
plt.title('Annual Rain Histogram')
plt.xlabel('Rain')
```

```python
# Rename crop yield columns
crop_yield_df.rename(columns={'Crop': 'Crop', 'Crop_Year': 'Year', 'Season': 'Season',
                            'State': 'State', 'Area': 'Area', 'Production': 'Production',
                            'Annual_Rainfall': 'Rainfall', 'Fertilizer': 'Fertilizer',
                            'Pesticide': 'Pesticide', 'Yield': 'Yield'}, inplace=True)

# Convert 'Year' to integer
crop_yield_df['Year'] = crop_yield_df['Year'].astype(int)

print("\nCleaned Crop Yield Data:")
print(crop_yield_df.head())
print(crop_yield_df.info())
```

```python
# Rename rain column
rain_df.rename(columns={'SUBDIVISION': 'State', 'YEAR': 'Year'}, inplace=True)

# Select relevant columns for annual rainfall calculation
monthly_rain_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

# Calculate annual rainfall
rain_df['Annual_Rainfall'] = rain_df[monthly_rain_cols].sum(axis=1)

# Select relevant columns
rain_df = rain_df[['State', 'Year', 'Annual_Rainfall']]

# Filter rain data for the overlapping period (1997-2017)
rain_df_filtered = rain_df[(rain_df['Year'] >= 1997) & (rain_df['Year'] <= 2017)].copy()

print("\nCleaned and Filtered Rain Data:")
print(rain_df_filtered.head())
print(rain_df_filtered.info())
```

```python
# Rename temperature columns
temp_df.rename(columns={'YEAR': 'Year', 'ANNUAL': 'Annual_Temperature',
                        'JAN-FEB': 'Jan_Feb_Avg', 'MAR-MAY': 'Mar_May_Avg',
                        'JUN-SEP': 'Jun_Sep_Avg', 'OCT-DEC': 'Oct_Dec_Avg'}, inplace=True)

# Filter temperature data for the overlapping period (1997-2021)
temp_df_filtered = temp_df[(temp_df['Year'] >= 1997) & (temp_df['Year'] <= 2021)].copy()

print("\nCleaned and Filtered Temperature Data:")
print(temp_df_filtered.head())
print(temp_df_filtered.info())
```

```python
# Now we will perform the analysis for the mean annual temperature. First we will check for missing values.
print("Missing values in temperature data:")
print(temp_df_filtered.isnull().sum())
```
